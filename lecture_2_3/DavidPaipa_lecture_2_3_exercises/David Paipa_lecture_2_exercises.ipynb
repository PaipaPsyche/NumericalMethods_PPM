### Write a program that determines the underflow and overflow limits

````python
import numpy as np

def underflow():
    base = 1;
    n= 1
    while(base/(2**(n+1))>0):
        n=n+1; # keep increasing the denominator exponent
    return base/(2**(n));
    

    
def overflow():
    base = 2;
    n= 1
    while(base**(n+1)>base**n):
        n=n+1; # keep increasing the denominator exponent
        print(n)
    return base**n;

#print("overflow found near the order of {}".format(overflow()))   ## keeps looping
print("underflow found near the order of {}".format(underflow())) 
    
````

### Write a program to determine the machine precision (within a factor of two)

````python
def machine_precision():
    n = -1
    while(1+10**(n-1)>1): ## while the new number is biggerthan the old one---
        n=n-1;
    return n;

print("Machine precision found in the order of 10^({})".format(machine_precision()-1))
````

### Explore subtractive cancellation

Consider the fucntions:
$$f(x)=\frac{1-\sqrt{1-t(x)^2}}{t(x)}$$  where $$t(x)=e^{-\pi x}$$

- plot the function $f(x)$ for $x$ in the range $[5,7]$

Rewrite the function $f(x)$ as

$$ \frac{t(x)}{1+\sqrt{1-t(x)^2}} $$

- plot on the same graph the two functions and comment

### Linear regression on simple data sets

- Write a function that given the input data returns the fitting parameters $a_1$ and $a_2$. The function template is:

```python
def lin_reg(x_data, y_data):
    """Function documentation"""

```

- Test your function on a simple set of data: plot the data and the fitted function

````python
def lin_reg(x_data, y_data):
    
    n = len(x_data) #total data
    x_mean = np.mean(x_data) #sum(x)/n
    y_mean = np.mean(y_data)#sum(y)/n
    
    s1 = np.sum([x_data[i]*y_data[i]-n*x_mean*y_mean for i in range(n)]) #S1 
    s2 = np.sum([x_data[i]*x_data[i]-n*x_mean*x_mean for i in range(n)]) #S2

    
    a1 = s1/s2 
    a0 = y_mean-a1*x_mean
    
    return (a0,a1)


lin_reg([0,1,2,3],[0,1,2,3])
````

## Bisection method

Use the bisection method to find the root of the function
    $$ f(x) = \frac{1}{2} - e^{-x}$$
think carefully how to estimate the error to end the calculation. 

````python
def condition_check(interval,f):
    return f(interval[0])*f(interval[1])<0
def bisection_method(interval,f,max_error=1E-15):
    
    itr = 0
    guess = (interval[1]+interval[0])/2 #first guess

    while(interval[1]-interval[0]>max_error): #while interval of search remains larger than error limit
        itr=itr+1
        mid =(interval[1]+interval[0])/2 # new guess
        last_guess=guess # unchanging case
        
        #intervals
        interval1 = [interval[0],mid] 
        interval2 = [mid,interval[1]]
        
        if(condition_check(interval1,f)): # if interval1 satisfies
            guess = (interval1[1]+interval1[0])/2
            interval = interval1
            
        elif(condition_check(interval2,f)): #else if interval2 satisfies
            guess = (interval2[1]+interval2[0])/2
            interval = interval2
        else: # no interval satisfies
            break
        
                    
        if(guess==last_guess):  #unchanging
            break
    print("solved in {} iterations".format(itr))
    return guess



def f_exponential(x):
    return 0.5 - np.exp(-x)


expected = np.log(2) #analytical solution
root_bisect = bisection_method([0,5],f_exponential) #root found with bisection method
err = abs(expected-root_bisect) #difference [ error ]

print("real = {}".format(expected))
print("bisection method = {}".format(root_bisect))
print("error = {}".format(err))
            
````

## False-position method

Use the false position method to find the root of the function
    $$ f(x) = \frac{1}{2} - e^{-x}$$
and compare to the bisection method

````python
def x_r_calc(interval,f):
    xu = interval[1]
    xl = interval[0]
    
    return xu - (f(xu)*(xl-xu))/(f(xl)-f(xu))


def false_position(interval,f,max_error=1E-15):
    itr = 0
    guess = x_r_calc(interval,f) #first guess

    while(interval[1]-interval[0]>max_error): #while interval of search remains larger than error limit
        
        itr=itr+1
        
        xr =x_r_calc(interval,f) # new guess
        last_guess=guess # unchanging case
        
        #intervals
        interval1 = [interval[0],xr] 
        interval2 = [xr,interval[1]]
        
        if(condition_check(interval1,f)): # if interval1 satisfies
            guess = x_r_calc(interval1,f)
            interval = interval1
            
        elif(condition_check(interval2,f)): #else if interval2 satisfies
            guess = x_r_calc(interval2,f)
            interval = interval2
        else: # no interval satisfies
            break
        
                    
        if(guess==last_guess):  #unchanging
            break
    print("solved in {} iterations".format(itr))
    return guess
    
false_p = false_position([0,5],f_exponential) #root found with bisection method
err = abs(expected-false_p) #difference [ error ]

print("real = {}".format(expected))
print("false position method = {}".format(false_p))
print("error = {}".format(err))   
````

This method find the root faster than the bisection method.

## The Secant Method

Implement the Newton-Rapshon method to solve 
$$ f(x) = \frac{1}{2} - e^{-x}$$
and compare to the bisection and false position methods

- Try different starting guess values, e.g. -1, 1, 5 and 30
- Comment

````python

def sec_method(initial,f,min_change=1E-15):
    
    itr=0
    
    x_prev = initial/2-min_change
    x = initial
    change = 10 #arbitrary value > min_change
    
    while(change>min_change): #while change is significant
        
        itr=itr+1
        
        new_x = x - f(x)*(x_prev-x)/(f(x_prev)-f(x)) 
        change = abs(x-new_x)
        x=new_x
    print("solved in {} iterations".format(itr))
    return x


def try_init(init):
    print("initial = {}".format(init))
    sec_val= sec_method(init,f_exponential) #root found with bisection method
    err = abs(expected-sec_val) #difference [ error ]

    print("real = {}".format(expected))
    print("secant method = {}".format(sec_val))
    print("error = {}".format(err)) 
    print("\n")
    
try_init(-5)
try_init(-1)
try_init(0)
try_init(1)
try_init(3)
try_init(4)
````

secant method is fast compared to the previous methods when the initial guess is near from the real root. When this guess is far from the expected value, convergence takes more steps, depending on the change(derivative) of the functio evaluated i nthe last guess. Function's derivative magnitude is huge far from the root towards the positive values, and small towards the negative ones so values bigger than 4 require long ties to converge, and values near -5 take longer but not as long as 5
