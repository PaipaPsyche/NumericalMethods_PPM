Implement the left, right, mid-point, trapezoidal and Simpson's rules and test them on a simple, analytical integral

````python
def Integral_left(f,interval,n):
        
        integral=0
        dx=abs((interval[0]-interval[1])/n)
        for i in range(n):
            area=f(interval[0]+i*dx)*dx
            integral+=area
            
          
            
        return integral

def Integral_right(f,interval,n):
        
        integral=0
        dx=abs((interval[0]-interval[1])/n)
        for i in range(1,n+1):
            area=f(interval[0]+i*dx)*dx
            integral+=area
            
          
            
        return integral
    

def Integral_mid(f,interval,n):
        
        integral=0
        dx=abs((interval[0]-interval[1])/n)
        xi=interval[0]
        for i in range(n):
            area=f(xi+dx/2)*dx
            xi+=dx
            integral+=area 
            
        return integral
    
def Integral_trap(f,interval,n):
    dx=abs(interval[0]-interval[1])/n
    xx=0
    for i in range(1,n+1):
        xx+=f(interval[0]+i*dx)
        
    integral=dx*((f(interval[0])+f(interval[1]))/2+xx)
    return integral

def Integral_simp(f,interval,n):
                xi=interval[0]
                xx=0
                dx=abs((interval[0]-interval[1])/n)
                integral1=0
                integral2=0
                for i in range(n):
                    area=f(xi+dx/2)*dx
                    xi+=dx
                    integral1+=area 
                for i in range(1,n+1):
                    xx+=f(interval[0]+i*dx)
                integral2=dx*((f(interval[0])+f(interval[1]))/2+xx)
                return (2*integral1+integral2)/3
        
        
    
````

## Monte Carlo Integration

In this exercise you are going to:

- implement a Monte Carlo integration routine 
- find the integral of $f(x)=x^2$ over the interval $[-2,2]$ 
- study the dependence of the absolute error on the number of samples.

Define the function $f(x) and a function `montecarlo` that returns the integral.

```python
def montecarlo(f,a,b,n):
    """f is function to integrate over the interval [a,b] with n random points"""
    ...
    ...
    return I
```

````python
import numpy as np
import matplotlib.pyplot as plt
import random

def montecarlo(f,interval,n):
    
    x=np.zeros(n)
    y=np.zeros(n)
    for i in range (n):
        x[i]=random.uniform(interval[0],interval[1])
        y[i]=f(x[i])
    I=(interval[1]-interval[0])*np.mean(y)
    return I

def square(x):
    return x*x
   
    
````

Calculate the integral and the absolute error, do the same integral with one of the methods you developed earlier. Compare and comment your results.

````python
print("The integral with Montecarlo algorithm is:",montecarlo(square,(-2,2),1000),"and the error is:", error(montecarlo(square,(-2,2),1000),16/3))
print("The integral with Simpson rule is:",Integral_simp(square,(-2,2),1000),"and the error is:", error(Integral_simp(square,(-2,2),1000),16/3))
````
